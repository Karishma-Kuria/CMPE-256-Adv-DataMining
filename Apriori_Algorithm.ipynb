# **Apriori Algorithm**

It is a algorithm for frequent item set mining and association rule learning for relational databases. 
It works in two steps


1.   Generating frequent itemsets: these are those itemsets which has
     transaction support above minimum support.   
2.   Generating all association rules from the frequent set generated in above
     step.

```python
# importing important libraries
import numpy as np
from apyori import apriori
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

## Loading the dataset from the drive

```python
ds_train = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/TRAIN-ARULES.csv')
ds_train.head()
```

Changing the format of the data in dataset

```python
# Converting comma seperation in product names to just space seperation
ds_train.product_name = ["".join(i.split(",")) for i in ds_train.product_name]

# Creating list of unique order_id's
orders_list = ds_train.order_id.unique()
orders_list
```

```python
# getting transactions for the unique order_id's
transactions = [ds_train[ds_train.order_id == i]["product_name"].tolist() for i in orders_list]
transactions
```

```python
transaction_encoder = TransactionEncoder().fit(transactions)
transactions = transaction_encoder.transform(transactions)
transactions = pd.DataFrame(transactions,columns=transaction_encoder.columns_)
transactions
```

## Calculating frequent itemset and Association rules using Apriori Algorithm

```python
# using Apriori with min_support = 0.0045
apriori_freq_itemset = apriori(transactions, min_support = 0.0045, use_colnames = True)
apriori_association_rules = association_rules(apriori_freq_itemset, metric = "confidence")
apriori_freq_itemset
```

## Generating Association rule using Apriori

```python
# association rules created using apriori
print("Association Rules generated using Apriori Method:- \n")
for item in apriori_association_rules.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
```
