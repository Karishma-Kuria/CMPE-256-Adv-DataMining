# **K Mean Cluster (Shopping Index & Income Index)**

```python
#importing required libraries
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
shoppingIndex = np.array([3, 3, 5, 2, 4, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
incomeIndex = np.array([5, 4, 6, 6, 5, 8, 2, 3, 6, 7, 2, 5, 1, 2, 6, 1, 3])

# plotting the dataset points on graph
labels = range(1,19)
plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Scatter Plot of Shopping Index & Income Index')
plt.xlabel('Shopping Index')
plt.ylabel('Income Index')
plt.scatter(shoppingIndex,incomeIndex)
plt.show()
```

# **Generic code to apply KMean Algorithm**

```python
# generic function to plot K-mean result
def applyKMean(array1, array2, color, marker):
  plt.plot()
  colors = color
  markers = marker
  plt.title('Scatter Plot of Shopping Index & Income Index')
  plt.xlabel('Shopping Index')
  plt.ylabel('Income Index')
  plt.plot()
  for i, l in enumerate(kmeans_model.labels_):
    plt.plot(array1[i], array2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])
  plt.show()
```

## **Applying k-mean algorithm for k=2**

```python
# applying k-mean algorithm for k=2
colors = ['b', 'g', 'r','y']
markers = ['o', 'v', 's','+']
K = 2
X = np.array(list(zip(shoppingIndex, incomeIndex))).reshape(len(shoppingIndex)
,2)
kmeans_model = KMeans(n_clusters=K).fit(X)
# print the cluster centroids
print(kmeans_model.cluster_centers_)
    
# get the cluster labels
print(kmeans_model.labels_)

# calling the generic function to Plot KMean result
applyKMean(shoppingIndex, incomeIndex, colors, markers)

# plotting the cluster centers and the data points on a 2D plane
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(shoppingIndex[i], incomeIndex[i], 
             color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    plt.scatter(kmeans_model.cluster_centers_[:, 0],
            kmeans_model.cluster_centers_[:, 1], 
            c='red', marker='x') 
    plt.title('Data points and the cluster centroids')
plt.show()
```
